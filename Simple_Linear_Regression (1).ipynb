Delivery_time -> Predict delivery time using sorting time

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import statsmodels.formula.api as smf
import warnings
warnings.filterwarnings("ignore")
```

```python
import pandas as pd
from google.colab import files
df=files.upload()
df=pd.read_csv("delivery_time.csv")
```

```python
df
```

```python
df.info()
```

```python
sns.distplot(df['Delivery Time'])
sns.distplot(df['Sorting Time'])
```

```python
#dataset=df.rename({'Delivery Time':'delivery_time'},{'Sorting Time':'sorting_time'},axis=1)
#dataset
df.corr()
```

```python
sns.set_style(style='darkgrid')
sns.pairplot(df)
```

```python
dataset=df.rename({'Delivery Time':'delivery_time','Sorting Time':'sorting_time'},axis=1)
dataset
```

```python
dataset.corr()
```

```python
sns.regplot(x=dataset['sorting_time'],y=dataset['delivery_time'])
```

```python
model=smf.ols("delivery_time~sorting_time",data=dataset).fit()
```

```python
model.summary()
```

```python
model.params
```

```python
model.tvalues
```

```python
model.pvalues
```

```python
model.rsquared
```

```python
model.rsquared_adj
```

Salary_hike -> Build a prediction model for Salary_hike


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import statsmodels.formula.api as smf
import warnings
warnings.filterwarnings("ignore")
```

```python
import pandas as pd
from google.colab import files
df=files.upload()
df=pd.read_csv("Salary_Data.csv")
```

```python
df.info()
```

```python
sns.distplot(df['YearsExperience'])
sns.distplot(df['Salary'])
```

```python
df.corr()
```

```python
sns.set_style(style='darkgrid')
sns.pairplot(df)
```

```python
dataset=df.rename({'YearsExperience':'Years_Experience','Salary':'salary'},axis=1)
dataset
```

```python
dataset.corr()
```

```python
sns.regplot(x=dataset['Years_Experience'],y=dataset['salary'])
```

```python
model=smf.ols("Years_Experience~salary",data=dataset).fit()
model.summary()
```

```python
model.params
```

```python
model.tvalues
```

```python
model.pvalues
```

```python
model.rsquared
```

```python
model.rsquared_adj
```
