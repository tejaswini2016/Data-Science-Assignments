```python
import numpy as np
data = np.array([24.23, 25.23, 25.41, 24.14, 29.62,28.25,25.81,24.39,40.26,32.95,91.36,25.99,39.42,26.71,35.00])
q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)
iqr = q3 - q1
threshold = 1.5 * iqr
outliers = np.where((data < q1 - threshold) | (data > q3 + threshold))

print("Outliers of array ",data,"is : \n", data[outliers])
```
